```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style #importing the libraries
```

```python
df = pd.read_csv("Total_monthly_sales.csv") #loading the dataset
df.head(3)
```

```python
#changing the columns titles so it is easier to understand
df.set_axis(["Month", "RIA", "RIA_LB", "RIA_SB", "REA", 
             "REA_LB", "REA_SB","PFS", "PFS_LB", "PFS_SB", 
             "NSFS", "NSFS_LB", "NSFS_SB", "SFS","AaOBaT", 
             "PNFS", "PNFS_LB", "PNFS_SB", "NSPNFS", "NSPNFS_LB",
             "NSPNFS_SB", "TCFL", "TCFL_LB", "TCFL_SB", "RST", 
             "RSC","RSC_LB", "RSC_SB", "FLG", "HGS", 
             "HGS_LB", "HGS_SB", "FL", "EHA", "HPG", 
             "AVaM", "ONFS","ONFS_LB", "ONFS_SB", "OR_PM", 
             "OR_BN", "OR_FC", "OR_CT", "OR_SSNEC","NSR", 
             "NSR_LB", "NSR_SB", "NS_MOH", "NS_EMOH", "AF"], axis="columns", inplace=True)
```

```python
df1.new = df1.apply(pd.to_numeric, errors='coerce')
```

```python
extracted_col = df["Month"]
display(extracted_col) #Joining the Month column back to the datatframe with the changed data types

df2 = df1.new.join(extracted_col)
df2.head()
```

```python
df2.dtypes #understanding the new data types
```

```python
df2 = df2.set_index('Month') #changing the axis to the month column
df2.head(2)
```
